```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import numpy as np
```

```python
key = []
dataframe = []
for f in glob.glob('/home/suraj/Repositories/Social-Cops/Cleaned Data/Jawali/*.csv'):
    key.append(f.split('/')[-1].split('.')[0].upper())
    data = pd.read_csv(f)
    del data["Unnamed: 0"]
    dataframe.append(data)
print key
result = pd.concat(dataframe, keys= key)
result
```

```python
def get_metric_data(series):
    label = str(series.iloc[1]) + str(series.name[0])
    x = []
    for d in range(2, 14):
        x.append(int(series.iloc[d]))
    return x, label

def subplot_bar(x, label, idx):
    """
    
    :param x: Range of values
    :param label: Metric definition
    :return: 
    """
    
    sns.plt.subplot(3,2,idx+1)
    sns.barplot(range(1, len(x)+1), x)
    sns.plt.title(label)

def plot_sum(x, label):
    sns.barplot(range(1, len(x)+1), x)
    sns.plt.title(label)
```

```python
x = result[result["Number"] == "1.4.1"]
x
```

```python
## Comparative Plot
frames = []
for idx in range(len(x)):
    y,label = get_metric_data(x.iloc[idx])
    subplot_bar(y, label, idx)
    frames.append(y)
```
