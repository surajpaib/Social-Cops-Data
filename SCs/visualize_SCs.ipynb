```python
% matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('dhamagaon.csv')
```

```python
data.head(10)
```

```python
def get_metric_data(series):
    label = str(series.iloc[:, 2].item())
    x = []
    for d in range(3, 15):
        x.append(int(series.iloc[:,d]))
    return x, label

def plot_bar(x, label):
    """
    
    :param x: Range of values
    :param label: Metric definition
    :return: 
    """
    sns.barplot(range(1, len(x)+1), x)
    sns.plt.title(label)
```

```python
x = data[data["Number"] == "1.1"]
y, label = get_metric_data(x)
plot_bar(y, label)
```
